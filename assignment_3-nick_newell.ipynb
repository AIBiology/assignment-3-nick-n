```python
x = df.drop(columns=['SalePrice', 'Street', 'Lot Frontage',
       'Year Built', 'BsmtFin SF 2', 'Central Air', 'Low Qual Fin SF', 'Gr Liv Area',
       'Kitchen AbvGr', 'Fireplaces', 'Garage Yr Blt',
       'Open Porch SF',
       'Enclosed Porch', 'Screen Porch', 'Yr Sold', 'Misc Val',
       'Mo Sold'])
y = df.SalePrice
#I started by only dropping SalePrice, but after running a few tests, I also dropped x-values w/ negative coefficient values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
#Changing the test_size from 0.25 to 0.2 improved the models score on testing data and RMSE
```

```python
model = LinearRegression() 
model.fit(x_train, y_train)
#I also tried using a pipeline for the model, but I couldn't get it to work, so I found information online that suggested using LinearRegression
```

```python
coeff_df = pd.DataFrame(model.coef_, x.columns, columns=['Coefficient'])
coeff_df
```

```python
y_pred = model.predict(x_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
print(f'Model score on training data: {model.score(x_train,y_train)}')
print(f'Model score on testing data: {model.score(x_test,y_test)}')
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
